```python
books_cleaned = books[['ISBN', 'Book-Title','Book-Author','Publisher']].copy()
```

```python
books_cleaned.columns = ['ISBN','Title','Author','Publisher']
```

```python
books_cleaned.dropna(inplace=True)# Missing values are being dropped
```

```python
books_cleaned.isnull().sum()
```

```python
books_cleaned.reset_index(drop=True, inplace=True)# Reset index
```

```python
users_cleaned = users.dropna()
```

```python
ratings_cleaned = ratings[ratings['ISBN'].isin(books_cleaned['ISBN'])].copy()
```

```python
crime_keywords = ['murder', 'crime', 'killer', 'detective', 'thriller', 'die', 'kill', 'investigation', 'homicide']
romance_keywords = ['love', 'romance', 'heart', 'passion', 'kiss']
fantasy_keywords = ['harry', 'magic', 'wizard', 'dragon', 'fantasy']

def assign_genre(title):
    title = title.lower()
    if any(word in title for word in fantasy_keywords):
        return 'Fantasy'
    elif any(word in title for word in crime_keywords):
        return 'Crime'
    elif any(word in title for word in romance_keywords):
        return 'Romance'
    else:
        return 'General'

books_cleaned['Genre'] = books_cleaned['Title'].apply(assign_genre)

```

```python
books_cleaned[books_cleaned['Title'].str.lower().str.contains("harry potter")]
```

```python
books_cleaned['combined_features'] = (
    books_cleaned['Title'].fillna('') + ' ' +
    books_cleaned['Author'].fillna('') + ' ' +
    books_cleaned['Genre'].fillna('')
)
```

```python
indices = pd.Series(books_cleaned.index, index=books_cleaned['Title'].str.lower())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(books_cleaned['combined_features'])
```

```python
from sklearn.neighbors import NearestNeighbors
nn_model = NearestNeighbors(n_neighbors=11, metric='cosine')
nn_model.fit(tfidf_matrix)
```

```python
import re
from difflib import get_close_matches
from tabulate import tabulate

def normalize_title(title):
    # Remove anything in parentheses (and spaces around) and lowercase the title
    return re.sub(r"\s*\(.*?\)\s*", "", title).strip().lower()

def recommend_books(query_title, n=10):
    query_title = query_title.lower().strip()

    
    if query_title in indices:
        idx = indices[query_title]
    else:
       
        keywords = query_title.split()
        possible_matches = books_cleaned['Title'].str.lower()
        partial_matches = possible_matches[possible_matches.apply(lambda t: all(k in t for k in keywords))]
        
        if not partial_matches.empty:
            idx = indices[partial_matches.iloc[0]]
        else:
            
            matches = get_close_matches(query_title, indices.index, n=1, cutoff=0.5)
            if matches:
                idx = indices[matches[0]]
            else:
                print(" Book not found in database.")
                return

    fetch_n = n * 5  
    distances, neighbors = nn_model.kneighbors(tfidf_matrix[idx], n_neighbors=fetch_n + 1)

    recommended = books_cleaned.iloc[neighbors[0][1:]][['Title', 'Author', 'Genre']].copy()

    recommended['norm_title'] = recommended['Title'].apply(normalize_title)
    recommended = recommended.drop_duplicates(subset='norm_title').drop(columns=['norm_title'])

    if recommended.empty:
        print(" No unique recommendations found after filtering duplicates.")
        return

    recommended = recommended.head(n)

    print(f"\n📚 Top {n} Recommendations for: **{books_cleaned.loc[idx, 'Title']}**\n")
    print(tabulate(recommended, headers='keys', tablefmt='fancy_grid', showindex=False))


   
```

```python
recommend_books("game of thrones")

```

# Collabrative-Based Filtering

```python
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
final_ratings =filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # Normalize the input
    book_name = book_name.strip().lower()
    
    # Normalize pt index
    normalized_index = pt.index.str.strip().str.lower()
    
    # Find all possible matches containing the input
    matching_books = [title for title in normalized_index if book_name in title]
    
    if not matching_books:
        print(f"No match found for '{book_name}'")
        return
    
    # Pick the first matching title (you can improve this later to pick the best)
    matched_book = matching_books[0]
    print(f"Showing results for: {matched_book}")

    index = np.where(normalized_index == matched_book)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
```

```python

recommend("lord of the rings")
```

```python
import re
from difflib import get_close_matches
from tabulate import tabulate

def normalize_title(title):
    return re.sub(r"\s*\(.*?\)\s*", "", title).strip().lower()

def hybrid_recommend_clean(book_query, n=10):
    book_query = book_query.strip().lower()
    
    # ---- Try Content-Based First ----
    content_found = False
    if book_query in indices:
        idx = indices[book_query]
        content_found = True
    else:
        keywords = book_query.split()
        possible_matches = books_cleaned['Title'].str.lower()
        partial_matches = possible_matches[possible_matches.apply(lambda t: all(k in t for k in keywords))]
        if not partial_matches.empty:
            idx = indices[partial_matches.iloc[0]]
            content_found = True
        else:
            matches = get_close_matches(book_query, indices.index, n=1, cutoff=0.5)
            if matches:
                idx = indices[matches[0]]
                content_found = True

    if content_found:
        fetch_n = n * 5
        distances, neighbors = nn_model.kneighbors(tfidf_matrix[idx], n_neighbors=fetch_n + 1)
        recommended = books_cleaned.iloc[neighbors[0][1:]][['Title', 'Author', 'Genre']].copy()
        recommended['norm_title'] = recommended['Title'].apply(normalize_title)
        recommended = recommended.drop_duplicates(subset='norm_title').drop(columns=['norm_title'])
        print(tabulate(recommended.head(n), headers='keys', tablefmt='fancy_grid', showindex=False))
        return  # ✅ Return early if content-based gave result

    # ---- Else Try Collaborative Filtering ----
    normalized_index = pt.index.str.strip().str.lower()
    matching_books = [title for title in normalized_index if book_query in title]

    if matching_books:
        matched_book = matching_books[0]
        index = np.where(normalized_index == matched_book)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:n+1]
        final = [[pt.index[i[0]]] for i in similar_items]
        print(tabulate(final, headers=["Title"], tablefmt="fancy_grid", showindex=False))
```

```python
hybrid_recommend_clean("harry potter")

```
